# 데이터 시각화의 개념 및 필요성

## 데이터 시각화
- 데이터의 분석 결과를 쉽게 이해하고 판단할 수 있도록 데이터를 시각적으로 표현하여 전달하기 위한 과정을 말한다.

## 필요성
- 많은 양의 데이터를 한눈에 파악할 수 있다. 시각적 요소를 활용해 데이터를 요약한다.
- 데이터 분석에 대한 전문적 지식이 없더라도 누구나 해당 데이터를 인지하고 활용할 수있게한다.
    - 사람이 감각기관을 통해 획득하는 정보의 80%는 시각에 의한 것이다.
    - 시각적인 입력은 다른 어떤 방법보다 빠르고 쉽게 사람이 이해할 수 있다.
    - 시각화 형태, 요소, 위치, 색등을 기반으로 패턴이나 다양한 인사이트(의미있는 정보)를 도출 및 표현할 수 있다.
- 단순한 데이터 요약, 통계분석 결과보다 정확한 데이터 분석결과를 도출할 수있다.
    - 데이터의 패턴이나 추세, 데이터간 비교등 표(Table)로 쉽게 인식 할 수 없는 것들을 시각적으로 쉽게 확인할 수 있다.

## matplotlib 그래프 구성요소


- **figure**
    - 전체 그래프가 위치할 기본 틀
    - 하나의 figure에 여러개의 그래프를 그릴 수 있다.
    - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
- **axes(subplot)**
    - figure내에 한개의 그래프를 그리기 위한 공간
        - figure에 한개 이상의 axes(subplot)로 구성해서 각 axes에 그래프를 그린다.
    - https://matplotlib.org/stable/api/axes_api.html
- **axis** 
    - 축 (x축, y축)은 값들을 위치시키는 선을 말한다.
    - axis label (x, y) : 축의 레이블(설명)
- **ticks** : 축의 값을 알려주는 눈금
    - Major tick
    - Minor tick
- **title** : 플롯 제목   
- **legend (범례)**
    - 하나의 axes내에 여러 그래프를 그린 경우 그것에 대한 설명

```python
%matplotlib inline
# %matplotlib qt

# matplotlib: 파이썬에서 데이터를 차트나 플롯(Plot)으로 시각화하는 패키지
# pyplot: matplotlib의 서브 패키지로서, 간단한 plot을 그릴 때 사용
import matplotlib.pyplot as plt
import numpy as np

# 한개의 figure에 여러개의 subplot 그리기
fig = plt.figure(figsize=(15,7), facecolor='gray') #facecolor: figure의 배경색
# 첫번쩨 subplot -> 1행 2열의 첫번째
axes1 = fig.add_subplot(1,2,1)
# 두번째 subplot -> 1행 2열의 두번째
axes2 = fig.add_subplot(1,2,2)

# plot 그리기 -> 선그래프
# plot(x축 데이터, y축 데이터, 선스타일, label='라벨')
axes1.plot([1,2,3,4,5], [10,20,30,40,50], label='line1')
axes1.plot([1,2,3,4,5], [50,40,30,20,10], label='line2')
# scatter 그리기 -> 산점도
# scatter(x축 데이터, y축 데이터, 마커모양, 마커색, label='라벨')
axes2.scatter(np.random.randint(100, size=50),
              np.random.randint(100, 200, size=50),
              color='r',
              label='scatter1')
axes2.scatter(np.random.randint(25, 75, size=50),
              np.random.randint(125, 175, size=50),
              color='b',
              label='scatter2')

# figure의 title 설정
# size: 폰트크기, color: 글자색
# 글자색은 영문 RGB(HEX)코드로 설정 또는 색상이름으로 설정 가능
fig.suptitle('Example of Plot', size=25, color='blue') 
# 각 plot의 title 설정
axes1.set_title("PLOT 1", size=20, color='#ffbbbb')
axes2.set_title("Plot 2", size=20, color='#bbffbb')

# 각 subplot의 tick(x축, y축)의 label 설정
axes1.set_xlabel("X축", size=15)
axes1.set_ylabel("Y축", size=15)
axes2.set_xlabel("가격1", size=15)
axes2.set_ylabel('가격2', size=15)

# 각 subplot의 legend 설정 (legend: 범례)
axes1.legend()
axes2.legend()
# subplot의 grid 설정 (grid: 격자)
axes1.grid(True)
# figure 출력
plt.show()
# # 주피터 노트북의 경우 코드의 마지막에 ;를 붙이면 출력됨
# ;
```

## pyplot 모듈을 이용해 그리기
1. pyplot 모듈이 그래프 그리는 함수와 Axes(Subplot) 설정 관련 함수를 제공

```python
# 그래프를 그릴때 사용할 데이터셋
x = [1,2,3,4,5]         # x축 데이터
y = [10,20,30,40,50]    # y축 데이터
```

```python
%matplotlib inline
# 1. import
import matplotlib.pyplot as plt
# figure의 크기 설정
# figsize=(가로크기, 세로크기) -> 단위: inch
plt.figure(figsize=(8,6))

# 2. plot 그리기
plt.plot(x, y)

# 3. 그래프에 대한 추가적인 설정
plt.title('예제')   # 그래프 제목
plt.xlabel('x축')   # x축 이름
plt.ylabel('y축')
plt.grid(True)      # 그래프에 grid(격자) 표시
plt.xticks(x)       # x축 눈금 라벨 설정 -> x축 데이터와 같게 설정
plt.yticks(y, labels=['가','나','다','라','마']) # y축 눈금 라벨 설정 -> 라벨을 직접 설정

# 4. 그래프 출력
plt.show()
```

2. 하나의 figure에 여러 그래프 그리기
    - plt.subplot(row,columns,num)

```python
# 하나의 axes(subplot)에 여러개의 그래프를 그리기
plt.bar(y+x,y+x, label='bar')
plt.plot(x+y,x+y, color='r', label='line')
plt.scatter(y+x,x+y, color='#00ff00', label='scatter')
plt.legend()
plt.show()
```

```python
# 1. import
import matplotlib.pyplot as plt
# figure의 크기 설정
plt.figure(figsize=(7,7))
# 각 그래프를 설정
# 위치 지정후, 그래프 그리기
plt.subplot(2,2,1)
plt.title('1번 그래프')
plt.xlabel('x축')
plt.ylabel('y축')
plt.plot(x, y)

plt.subplot(2,2,2)
plt.title('2번 그래프')
plt.xlabel('x축')
plt.ylabel('y축')
# scatter 그래프 그리기 -> scatter: 산점도를 그릴 때 사용
plt.scatter(x, y)

plt.subplot(2,2,3)
plt.title('3번 그래프')
# bar 그래프 그리기 -> bar: 양적인 비교를 할 때 사용
plt.bar(x, y)

plt.subplot(2,2,4)
plt.title('3번 그래프')
# pie 그래프 그리기 -> pie: 비율을 나타낼 때 사용
plt.pie(x, autopct='%.2f') 

plt.tight_layout() # 그래프간의 간격을 자동으로 맞춤
plt.show()
```

## Figure 와 Axes 객체를 이용해 그리기

- AxesSubplot 객체(Axes를 표현하는 객체)의 메소드들을 이용해 그래프를 그린다.
- axes 생성 방법
    - plt.gca(): AxesSubplot
        - 하나의 figure에 하나의 Axes 만 사용할 경우 사용.
    - 하나의 figure에 여러개의 Axes들을 사용할 경우
        - figure.add_subplot() 메소드 이용
            - figure를 먼저 생성후 axes 들을 추가
        - pyplot.subplots() 함수를 이용
            - figure와 axes배열을 동시에 생성

### Figure에 하나의 axes 그리기
- plt.gca() 사용

```python
# 1. import
import matplotlib.pyplot as plt
# figure의 크기 설정
plt.figure(figsize=(3,3))

ax = plt.gca() # 현재 axes를 얻어옴
ax.plot(x, y, label='line1')
ax.plot(x, y[::-1], label='line2')
ax.set_title('1번')
ax.grid(True, linestyle='--', linewidth=1, alpha=0.5, color='#ff0084') # grid 설정
plt.show()
```

### 하나의 Figure에 여러개의 axes 그리기
1. <b style='font-size:1.3em'>figure.add_subplot() 메소드 이용</b>
    - figure객체에 axes를 추가하는 형태
    - nrows(총행수), ncols(총열수), index(axes위치) 지정

```python
# 1. import
import matplotlib.pyplot as plt
# figure의 크기 설정
fig = plt.figure(figsize=(5,5))

# 2. subplot들의 위치를 설정
# 전체 figure의 행, 열갯수와 몇번째 subplot인지를 설정
ax1 = fig.add_subplot(2,2,1) # 2행 2열의 첫번째 subplot
ax2 = fig.add_subplot(1,2,2) # 1행 2열의 두번째 subplot
ax3 = fig.add_subplot(223) # 2행 2열의 세번째 subplot

ax1.set_title('1번')
ax1.plot(x, y)
ax2.set_title('2번')
ax2.scatter(x, y)
ax3.set_title('3번')
ax3.bar(x, y)
ax4.set_title('4번')
ax4.pie(x+y, autopct='%.2f')

plt.tight_layout()
plt.show()
```

2. <b style='font-size:1.3em'>pyplot.subplots()</b>
    - nrows, ncols 로 axes개수와 위치 지정
    - 반환: figure와 axes(subplot)들을 담은 **ndarray**
