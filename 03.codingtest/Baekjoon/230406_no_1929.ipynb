```python
#  
prime_num = [2, 3, 5, 7]

#  
def eratosthenes_shecker(num : int) -> None:
    """
          
    Args:
        num (type[int]):       
        
        from_is_prime (bool, optional): num is_prime   . Defaults to True.
    """
    n_min = 2
    
    #        
    n_max = int(num**0.5)+1
        
    #      
    if prime_num[-1]+1 % 2 == 0 :
        prime_num.extend(list(range(prime_num[-1]+1,n_max))[0::2]) 
    else :
        prime_num.extend(list(range(prime_num[-1]+1,n_max))[1::2])
    
    
    # ' ' 
    for i in prime_num:
        for j in prime_num:
            if i!=j and 0==(j%i):
                prime_num.remove(j)
    return

#  
def is_prime(num : int) -> bool :
    """
      
    Args:
        num (type[int]):    

    Returns:
        bool:  True,  False
    """
    result = True
    
    #     
    if num in prime_num :
        return result

    #           
    if num > prime_num[-1] :
        eratosthenes_shecker(num)
    
    #        
    for v in prime_num :
        if num % v == 0 :
            result = False
    
    return result
```

```python
# map    
# n, m = map(int, input().split(' '))
n, m = map(int, test_input.split(' '))

# n m  
for i in range(n, m+1):
    if is_prime(i) :
        print(i)
```
